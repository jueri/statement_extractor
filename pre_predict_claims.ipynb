# Prepredict claims
Notebook to predict claim sentence probabilitys on a sentence table.

```python
import csv

import pandas as pd

from src import detect_claims
from config import MODEL_NAME, MODEL_WEIGHTS_PATH
```

```python
data = pd.read_csv("SMC_claim_sentences.csv", lineterminator='\n')
```

```python
detector = detect_claims.claim_detector(MODEL_NAME, MODEL_WEIGHTS_PATH)
```

```python
with open("SMC_claim_sentences_pre_pred.csv", "a+") as outfile:
    writer = csv.writer(outfile)
    c = 0
    for i in data.iterrows():
        if i[0] <= c:
            continue
            
        prob = detector.classify(i[1].sentence)[0][1]
        writer.writerow([i[1].sentence_ID, i[1].title, i[1].sentence, prob])
        print(c, "/", len(data))
        c+=1
```
